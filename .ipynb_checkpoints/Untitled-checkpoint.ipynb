```python
import pandas as pd
```

```python
df = pd.read_csv('Viviendas09.CSV')
df
```

```python
df = df.drop(columns=['ENT','LOC50K','COBERTURA','ESTRATO','UPM','FACTOR','LUG_COC','COMBUSTIBLE','ESTUFA','FOCOS','FOCOS_AHORRA','ABA_AGUA_NO_ENTU','BOMBA_AGUA','REGADERA','SEPARACION1','SEPARACION2','SEPARACION3','SEPARACION4','DESTINO_BAS','TENENCIA','ESCRITURAS','FORMA_ADQUI','FINANCIAMIENTO1','FINANCIAMIENTO2','FINANCIAMIENTO3','DEUDA','DUE1_NUM','DUE2_NUM','MCONMIG','INGR_PEROTROPAIS','INGR_PERDENTPAIS','INGR_AYUGOB','INGR_JUBPEN','ALIMENTACION','ALIM_ADL1','ALIM_ADL2','ING_ALIM_ADL1','ING_ALIM_ADL2','ING_ALIM_ADL3','INGTRHOG','TAMLOC'])
```

```python
df = df.dropna(axis=0, how='all')
df
```

```python
convertir = {
    1: "Casa única en el terreno",
    2: 'Casa que comparte terreno con otra(s)',
    3: 'Casa dúplex',
    4: 'Departamento en edificio',
    5: 'Vivienda en vecindad o cuartería',
    6: 'Vivienda en cuarto de azotea de un edificio',
    7: 'Local no construido para habitación',
    8: 'Vivienda móvil',
    9: 'Refugio',
    99: 'No especificado de vivienda particular'
}
```

```python
df['CLAVIVP'] = df['CLAVIVP'].map(convertir)
```

```python

def cuarto(value):
    if(value==99): return "No especificado"
    else: return value
```

```python
df['CUADORM'].map(cuarto)
```

```python
df.groupby('MUN')['CLAVIVP'].value_counts()
```
